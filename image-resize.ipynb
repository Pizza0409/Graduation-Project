# Resize image

```python
from PIL import Image
import os

# Specify the directory path
directory_path = '/kaggle/input/test-data/p1'

# Specify the directory for saving resized images
output_directory = '/kaggle/working/test_data_resize'

# Loop through all files in the directory
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.lower().endswith((".jpg", ".png")):
            # Construct the full file path
            image_path = os.path.join(root, file)

            # Open the image
            img = Image.open(image_path)
            
            # Get the original size
            (w, h) = img.size
            print(f'Original size - w={w}, h={h}')

            # Resize the image to 640x640
            new_img = img.resize((640, 640))
            new_w, new_h = new_img.size
            print(f'Resize image - w={new_w}, h={new_h}')
            # Save the resized image with the correct format
            base_name, ext = os.path.splitext(file)
            new_image_path = os.path.join(output_directory, f"{base_name}-640x640{ext}")
            new_img.save(new_image_path)
```
